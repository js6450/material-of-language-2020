In this assignment, I've decided to take iMessage data of my conversations with 4 different people and turn them into images. I wanted to see if my conversation patterns with these different people would be more visible by turning text data into grayscale images.

4 files were prepared per person and each are collection of text data that came in the format of:

```
(timestamp) Name: Text message content
```

First part of the code below opens the text file, splits by newline character, then parses through the text to only get the text message content and concatenates it into a string variable.

````python
inputOne = open("a.txt").read()

lines = inputOne.split('\n')

cleanLines = ""

for line in lines:
    if ")" in line:
        firstSplit = line.split(')')
        secondSplit = firstSplit[1].split(":")

        if len(secondSplit) > 1:
            cleaned = secondSplit[1].strip()
            
            if len(cleaned) > 0 and ord(cleaned[0]) != 65532:
                cleanLines += cleaned
                cleanLines += " "

aCleaned = str.encode(cleanLines[-1001:].strip())

len(aCleaned)
````

Then I am importing the numpy library to create a blank array to use as pixel data for im image. I'm also importing the Pilow library.

````python
import numpy as np
from PIL import Image
````

I'm creating a vertical image, where each column represents a byte data of the text conversation with this particular person, where the byte value is being used as the alpha value.

````python
data = np.zeros((1000, 1000, 4), dtype=np.uint8)

for i in range(1000):
    for j in range(1000):
        data[j, i] = [0, 0, 0, aCleaned[i]]

aImg = Image.fromarray(data)
aImg
````
